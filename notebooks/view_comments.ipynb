Get table comments. First select gets table definitions, 2nd select gets column definions

```python
QUERY = """
select
  pg_namespace.nspname as schema_name,
  pg_class.relname as table_name,
  c.column_name,
  pg_description.description
from pg_description
  join pg_class
    on pg_description.objoid = pg_class.oid
  join pg_namespace
    on pg_class.relnamespace = pg_namespace.oid
  left join information_schema.columns as c
    on pg_description.objsubid = c.ordinal_position
    and c.table_schema = pg_namespace.nspname
    and c.table_name = pg_class.relname
where c.column_name is null


union


select
  c.table_schema
  ,c.table_name
  ,coalesce(c.column_name,'') as column_name
  ,pg_description.description
from pg_description
  join pg_class
    on pg_description.objoid = pg_class.oid
  join pg_namespace
    on pg_class.relnamespace = pg_namespace.oid
  right join information_schema.columns as c
    on pg_description.objsubid = c.ordinal_position
    and c.table_schema = pg_namespace.nspname
    and c.table_name = pg_class.relname
where c.table_schema not in ('information_schema', 'pg_catalog', 'public')
order by 1,2,3 nulls first
        
"""



with engine.begin() as conn:
    df = pd.read_sql(QUERY, conn)
```
